```python
import random
```

```python
card_list = []
suits = ['Hearts','Spades','Clubs','Diamonds']
face_cards = ['King', 'Queen', 'Jack', 'Ace']

for i in suits:
    for j in range(2,11):
        card_list.append('{} of {}'.format(j,i))
        
for i in suits:
    for j in face_cards:
        card_list.append('{} of {}'.format(j,i))
        
card_value = {}
temp_list = []

for i in range (36):
    temp_list.append(card_list[i])

temp_value = []    

for i in range(4):
    for j in range(2,11):
        temp_value.append(j)

for i in range(36):
    card_value.update({temp_list[i]:temp_value[i]})
    
all_face_cards = {'Jack of Hearts':10,'Queen of Hearts':10,'King of Hearts':10,'Ace of Hearts':11,
                   'Jack of Spades':10,'Queen of Spades':10,'King of Spades':10,'Ace of Spades':11,
                   'Jack of Clubs':10,'Queen of Clubs':10,'King of Clubs':10,'Ace of Clubs':11,
                   'Jack of Diamonds':10,'Queen of Diamonds':10,'King of Diamonds':10,'Ace of Diamonds':11}

card_value.update(all_face_cards)
```

```python
class Money():
    
    def __init__(self,balance):
        
        self.balance = balance
        
    def deposit(self,amount):
        
        self.balance += amount
        
        print('Current Balance after Deposit = {}'.format(self.balance))
        
    def withdraw(self,amount):
        
        if self.balance >= amount:
            
            self.balance -= amount
            
            print('Current Balance after Withdrawal = {}'.format(self.balance))
            
        else:
            print('Insufficient Balance')
            
player = Money(1000)
```

```python
print('Welcome to the BLACKJACK game!')

while (player.balance>0) and (card_list):

    def betting_amount():

        while True:

            print(f'Your current balance is {player.balance}...')

            bet = int(input('Please enter your betting amount: '))

            if bet > player.balance:
                continue
            else:
                break

        return bet

    bet_amount = betting_amount()


    random.shuffle(card_list)

    player_cards = []
    dealer_cards = []

    for i in range(2):
        player_cards.append(card_list.pop())
        dealer_cards.append(card_list.pop())

    print('The Players cards are: {} and {}'.format(player_cards[0],player_cards[1]))
    print('The dealers card is: {}'.format(dealer_cards[0]))

    current_player_total = card_value[player_cards[0]] + card_value[player_cards[1]]

    if current_player_total > 21:
        print('Your score has exceeded 21, Hence you Loose')
        player.withdraw(bet_amount)


    else:

        while current_player_total <= 21:

            hit = input('Do you want to HIT? (Y/N): ').upper()

            if hit == 'Y':
                player_cards.append(card_list.pop())

                current_player_total += card_value[player_cards[-1]]

                print('The Players card is: {}'.format(player_cards[-1]))

                if current_player_total > 21:

                    print(f'Player total = {current_player_total}')

                    print('Your score has exceeded 21, Hence you Loose')

                    player.withdraw(bet_amount)


                continue

            else:

                print(f'Player total = {current_player_total}')

                break


    if current_player_total <= 21:

        print('The dealers card is: {}'.format(dealer_cards[1]))

        current_dealer_total = card_value[dealer_cards[0]] + card_value[dealer_cards[1]]

        if current_dealer_total > 21:
            print("Dealer's score has exceeded 21, Hence you WIN!!")

            player.deposit(bet_amount)


        else:

            while current_dealer_total < 17:

                dealer_cards.append(card_list.pop())

                current_dealer_total += card_value[dealer_cards[-1]]

                print('The Dealer card is: {}'.format(dealer_cards[-1]))

                if current_dealer_total > 21:
                    print("Dealer's score has exceeded 21, Hence you WIN!!")

                    player.deposit(bet_amount)

                    break

            if current_dealer_total <=21:

                if current_dealer_total > current_player_total:

                    print(f'Dealer total = {current_dealer_total}')
                    print('Dealer has more score than the Player, Hence the Dealer wins!')

                    player.withdraw(bet_amount)

                else:
                    print(f'Dealer total = {current_dealer_total}')
                    print('Player has more score than the Dealer, Hence the Player wins!')

                    player.deposit(bet_amount)




    if player.balance > 0:
        
        game_on = input('Do you want to continue playing? (Y/N): ').upper()

        if game_on == 'Y':
            continue

        else:
            break
```
