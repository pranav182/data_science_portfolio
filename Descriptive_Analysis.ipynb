# Project for Descriptive Statistics
In this project, we shall import a dataset about cars and try to observe some of the measures. Some of the tasks we shall perform are: calculating mean, median, variance, standard deviation, plotting histograms, density plots, box plots and drawing inferences.

These are the columns present in mtcars dataset:

- name - Name of the car;
- mpg - Miles/(US) gallon; 
- cyl - Number of cylinders;
- disp - Displacement (cu.in.);
- hp - Gross horsepower;
- drat - Rear axle ratio;
- wt - Weight (lb/1000);
- qsec - 1/4 mile time;
- vs - V/S;
- am - transmission (0 = automatic, 1 = manual);
- gear - Number of forward gears;
- carb - Number of carburetors.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from ggplot import mtcars
```

```python
print(mtcars)
```

```python
header_data = mtcars.columns.values
```

```python
print(header_data)
```

```python
type(header_data)
```

```python
# Calculate the mean of each of the columns of mtcars.

print(mtcars.mean())
```

```python
# Get the mean of each row (car) and store it in variable 'mtcarsMeanCarWise'.

mtcarsMeanCarWise = mtcars.mean(axis=1)
print(mtcarsMeanCarWise)
```

```python
# Get the median of each column and store it in variable 'medianCarsFeatures'

medianCarsFeatures = mtcars.median()
print(medianCarsFeatures)
```

```python
# Plot the histograms for each of the columns in mtcars using hist() function with bin-size as 10.

mtcars.hist(bins = 10, figsize=(20,15))
plt.show()
```

```python
# Plot the density curves for each column of mtcars using 'for' loop and plot command.

for i in range(1, len(header_data)):
    print(header_data[i])
    mtcars[header_data[i]].plot.density()
    plt.show()
```

```python
# Plot the correlation for the following attributes: mpg, disp, hp, wt, qsec. You can use the scatter_matrix command present in pandas.plotting library.

from pandas.plotting import scatter_matrix

attributes = ["mpg", "disp", "hp", "wt", "qsec"]

scatter_matrix(mtcars[attributes], figsize = (12,8))
```

```python
# We use skewness as a measure of symmetry. If the skewness of is zero 
# then the distribution represented is perfectly symmetric. If the skewness
# is negative, then the distribution is skewed to the left, while if 
# the skew is positive then the distribution is skewed to the right. 
# Hence, it can also be considered as a measure of normality since normal 
# distributions are symmetric and hence have a zero skew. Calculate the 
# skeweness of each column of 'mtcars' using mtcars.skew()

print(mtcars.skew())  
```

```python
mtcars.describe()
```

```python
# Use describe and quantile to calculate the Inter-Quantile spread of 'mpg' 
# attribute of mtcars.

print(mtcars["mpg"].quantile())
```

```python
# Plot the box plot for 'mpg' column of mtcars.

print(plt.boxplot(mtcars["mpg"]))
```

```python
# Calculate the mean, standard deviation and variance of the 'mpg' column

print("Mean = ", mtcars["mpg"].mean())
print("Standard Deviation = ", mtcars["mpg"].std())
print("Variance = ", mtcars["mpg"].var())
```

```python
# Extract cars with automatic transmission and cars with manual 
# transmission into separate variables 'mtCarsAutomatic' and 
# 'mtCarsManual' respectively

mtCarsAutomatic = mtcars[mtcars['am'] == 0]

mtCarsManual = mtcars[mtcars['am'] == 1]
```

```python
print(mtCarsAutomatic)
```

```python
print(mtCarsManual)
```

```python
# Plot the box plot for 'mpg' (miles per gallons) for each of the subsets 
# of cars grouped by transmission (automatic/ manual).

print(plt.boxplot(mtCarsManual["mpg"]))
```

```python
print(plt.boxplot(mtCarsAutomatic["mpg"]))
```

# Density Curves

```python
# Generate 10000 random normally distributed numbers and plot the density
# curve along with the median and mean.

norm_data = pd.DataFrame(np.random.normal(size = 100000))
```

```python
norm_data.plot.density()
plt.show()
```

```python
print(norm_data.mean())
```

```python
print(norm_data.median())
```

```python
# Generate 10000 exponentially distributed numbers using np.random.exponential
# (size=100000) and plot the density curve along with the median and mean.

skewed_data = pd.DataFrame(np.random.exponential(size = 100000))
```

```python
skewed_data.plot(kind="density",figsize=(10,10),xlim=(-1,5))
```

```python
print(skewed_data.mean())
```

```python
print(skewed_data.median())
```
